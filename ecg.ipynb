```python
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
d = ET.parse("Example.xml")
r = d.getroot()
f = r.find('{urn:hl7-org:v3}component')
for i in range (1, 13):
    t = f[0][8][0][i][0][1][2]
    result = t.text
    data = list(map(int, result.split()))
    cord = [i for i in range (0, len(data))]
    g = f[0][8][0][i][0][0].attrib
    label = g.get('code')
    print(label)
    plt.figure(figsize=(5,5))
    plt.plot(cord, data)
    plt.show()
```
